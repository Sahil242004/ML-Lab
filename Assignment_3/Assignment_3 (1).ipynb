```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv')
df = pd.DataFrame(data)
df.head(10)
# Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.
```

```python
y = df['Chance of Admit ']
x = df['CGPA']
print(x)
```

Scatter plot

```python
plt.scatter(x,y)
plt.show()
```

```python
x = df['GRE Score']
y = df['Chance of Admit ']
plt.scatter(x,y,color='red')
```

```python
x = df['GRE Score']
y = df['TOEFL Score']
plt.scatter(x,y,color='red')
```

```python
ax = sns.scatterplot(x='GRE Score',y='TOEFL Score', data=df)
ax.set(xlabel="GRE Score", ylabel="TOEFL Score", title="GRE vs. TOEFL")
```

Bar plot


```python
# bar plot
y = df['Chance of Admit ']
x = df['CGPA']
plt.xlabel('CGPA')
plt.ylabel('Chance of Admit')
plt.title("Chance of Admit vs CGPA")
plt.bar(x,y)
plt.show()
```

```python
x = df['GRE Score']
y = df['Chance of Admit ']
plt.xlabel('GRE Score')
plt.ylabel('Chance of Admit')
plt.title("Chance of Admit vs CGPA")
plt.bar(x,y)
plt.show()
```

```python
x = df['GRE Score']
y = df['TOEFL Score']
plt.xlabel('GRE Score')
plt.ylabel('TOEFL Score')
plt.title("Chance of Admit vs CGPA")
plt.bar(x,y)
plt.show()
```

```python
sns.barplot(x='GRE Score',y='TOEFL Score', data=df, palette="coolwarm")
plt.xticks(rotation=90, ha='right')
plt.show()
```

Box Plot

```python
y = df['Chance of Admit ']
x = df['CGPA']
plt.boxplot(y , vert=False)
```
